## What is ColumnTransformer?
**ColumnTransformer is a powerful tool from scikit-learn that allows you to apply different preprocessing techniques to different columns of your dataset — all in a single step**.

## Why do we use it?
- Real-world datasets often contain a mix of:
- Numerical features (e.g., age, income)
- Categorical features (e.g., gender, country)
- Text features
- Date/time features
**These require different preprocessing steps (e.g., scaling for numbers, one-hot encoding for categories).
Without ColumnTransformer, you'd need to process each column separately, making your code messy and error-prone.**


```python
import pandas as pd 
import numpy as np 
```

```python
from sklearn.impute  import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder 
```

```python
df=pd.read_csv(r"C:\Users\Asus\Downloads\covid_toy.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test =train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                               test_size=0.2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer= ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.transform(X_test).shape
```
